```python
#trying some code to calculate scores
from matchms import Spectrum
from matchms.similarity import CosineGreedy
from matchms.filtering import default_filters
from matchms.importing import load_from_mgf
from matchms import calculate_scores
import numpy as np

def within_tolerance(reference_mz, query_mz, tolerance_ppm=20):
    """Check if the reference mass is within the specified tolerance of the query mass."""
    tolerance = query_mz * (tolerance_ppm / 1e6)
    return abs(reference_mz - query_mz) <= tolerance
```

```python
def filter_spectra_by_precursor(reference_spectra, query_spectrum):
    """Filter spectra by precursor mass within 20 ppm tolerance and return both the spectra and their indices."""
    query_precursor_mz = query_spectrum.get("precursor_mz")
    filtered_spectra = []
    indices = []
    for i, spectrum in enumerate(reference_spectra):
        if within_tolerance(spectrum.get("precursor_mz"), query_precursor_mz):
            filtered_spectra.append(spectrum)
            indices.append(i)
    return filtered_spectra, indices

def compute_similarity_scores(query_spectrum, reference_spectra):
    similarity_measure = CosineGreedy(tolerance=0.02)
    scores = calculate_scores(reference_spectra, [query_spectrum], similarity_measure, is_symmetric=False)
    scores_array = scores.scores.to_array()  # Convert scores to structured array = 
    return scores_array
```

```python
def get_filtered_scores_and_indices(scores_array, filtered_indices):
    """Filter scores based on cosine score and number of matches, return original indices."""
    results = []
    # Assuming 'scores' is a structured array with multiple rows, each representing a score from a spectrum comparison
    for i in range(len(scores_array)):
        cos_score = scores_array[i]['CosineGreedy_score']
        num_matches = scores_array[i]['CosineGreedy_matches']
        if cos_score >= 0.4 and num_matches >= 4:
            original_index = filtered_indices[i]  # Get the original index from filtered indices
            results.append([original_index, cos_score, num_matches])
    return results
```

```python
##Trying to compare all spectrums in the spectra
from matchms.importing import load_from_mgf
from matchms.similarity import CosineGreedy
from matchms import calculate_scores
import numpy as np

def within_tolerance(reference_mz, query_mz, tolerance_ppm=20):
    """Check if the reference mass is within the specified tolerance of the query mass."""
    tolerance = query_mz * (tolerance_ppm / 1e6)
    return abs(reference_mz - query_mz) <= tolerance

def filter_spectra_by_precursor(reference_spectra, query_spectrum):
    """Filter spectra by precursor mass within 20 ppm tolerance and return both the spectra and their indices."""
    query_precursor_mz = query_spectrum.get("precursor_mz")
    filtered_spectra = []
    indices = []
    for i, spectrum in enumerate(reference_spectra):
        if within_tolerance(spectrum.get("precursor_mz"), query_precursor_mz):
            filtered_spectra.append(spectrum)
            indices.append(i)
    return filtered_spectra, indices
```

```python
def compute_similarity_scores(query_spectrum, reference_spectra):
    """Compute cosine similarity scores using the CosineGreedy method."""
    print("Number of reference spectra:", len(reference_spectra))
    # print("1. reference_spectra: ", reference_spectra[:2])
    if not reference_spectra:  # Check if the reference spectra list is empty
        # return np.array([(0, 0, 0)], dtype=[('index', 'i4'), ('CosineGreedy_score', 'f4'), ('CosineGreedy_matches', 'i4')])
        return np.array([], dtype=[('index', 'i4'), ('CosineGreedy_score', 'f4'), ('CosineGreedy_matches', 'i4')])
    similarity_measure = CosineGreedy(tolerance=0.02)
    
    # print("2.1 query_spectrum", type(query_spectrum.mz))
    
    scores = calculate_scores(reference_spectra, [query_spectrum], similarity_measure, array_type = "sparse", is_symmetric=False)
    print("3. scores: ", scores)
    scores_array = scores.scores.to_array()
    
    return scores_array

def get_filtered_scores_and_indices(scores_array, filtered_indices):
    """Filter scores and return the highest cosine score result if multiple meet the criteria."""
    best_result = None
    highest_cos_score = 0
    for i in range(len(scores_array)):
        cos_score = scores_array[i]['CosineGreedy_score']
        num_matches = scores_array[i]['CosineGreedy_matches']
        if cos_score >= 0.6 and num_matches >= 6:
            if cos_score > highest_cos_score:
                highest_cos_score = cos_score
                original_index = filtered_indices[i]
               # Convert numpy arrays to scalar if they are numpy arrays
                cos_score_scalar = cos_score.item() if isinstance(cos_score, np.ndarray) else cos_score
                num_matches_scalar = num_matches.item() if isinstance(num_matches, np.ndarray) else num_matches
                best_result = [original_index, cos_score_scalar, num_matches_scalar]
    if best_result is None:
        best_result = [0, 0, 0]
    return [best_result]
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from pyteomics import mzml

def calculate_average_peaks_per_spectrum(mzml_files):
    avg_peaks_per_file = {}

    for file_mzml in mzml_files:
        total_peaks = 0
        ms1_count = 0
        
        with mzml.read(file_mzml) as reader:
            for spectrum in reader:
                ms_level = spectrum.get('ms level')
                
                # Only process MS1 spectra (ms level = 1)
                if ms_level == 1:
                    mz_array = spectrum.get('m/z array')
                    total_peaks += len(mz_array)
                    ms1_count += 1
            
            # Calculate average number of peaks per MS1 spectrum for this file
            avg_peaks = total_peaks / ms1_count if ms1_count > 0 else 0
            # Use only the last part of the file name (before ".mzML")
            file_label = os.path.basename(file_mzml).replace('.mzML', '')
            avg_peaks_per_file[file_label] = avg_peaks

    return avg_peaks_per_file
```
